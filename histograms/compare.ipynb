```python
import numpy as np
from matplotlib import pyplot as plt
from tools import cdffit, pdffit
import seaborn
from astropy.stats import bayesian_blocks as bb
```

# Normal distribution

```python
x = np.random.normal(size=100)
```

```python
fig = plt.figure(dpi=150, figsize=(8, 4))
ax = fig.add_subplot(141)
ax.hist(x, density=True)
ax.set_title('Histogram')

ax = fig.add_subplot(142)
edges = bb(x, fitness='events', p0=0.05)
plt.hist(x, bins=edges, density=True)
ax.set_title('Bayesian blocks')

ax = fig.add_subplot(143)
seaborn.kdeplot(x, ax=ax)
ax.set_title('KDE')

ax = fig.add_subplot(144)
x.sort()
y = np.linspace(0, 1, len(x))
ref = (x-x.min())/(x.max()-x.min())
res = y - ref
pcdf = y.copy()
pcdf[y>0.5] = 1 - pcdf[y>0.5]
ax.set_title('ECDF')
ax.plot(x, y)
```

# A more complicated distribution

```python
from scipy.stats import gamma as g

g1 = g(a=3, loc=0, scale=0.8)
g2 = g(a=2, loc=0.5, scale=0.05)
frac1 = 0.8
N = 300
N1 = int(N*frac1)
N2 = N - N1
r1 = g1.rvs(size=N1)
r2 = g2.rvs(size=N2)
x = np.concatenate([r1, r2])
```

```python
fig = plt.figure(dpi=150, figsize=(12, 4))
ax = fig.add_subplot(141)
t = np.linspace(0, 10, 200)
G = frac1*g1.pdf(t) + (1-frac1)*g2.pdf(t)
ax.plot(t, G)
```

```python
fig = plt.figure(dpi=150, figsize=(8, 4))
ax = fig.add_subplot(141)
ax.hist(x, density=True)
ax.set_title('Histogram')

ax = fig.add_subplot(142)
edges = bb(x, fitness='events', p0=0.01)
plt.hist(x, bins=edges)
ax.set_title('Bayesian blocks')

ax = fig.add_subplot(143)
x.sort()
y = np.linspace(0, 1, len(x))
ref = (x-x.min())/(x.max()-x.min())
res = y - ref
pcdf = y.copy()
pcdf[y>0.5] = 1 - pcdf[y>0.5]
ax.set_title('ECDF')
ax.plot(x, y)

ax = fig.add_subplot(144)
seaborn.kdeplot(x, ax=ax)
ax.set_title('KDE')
```
